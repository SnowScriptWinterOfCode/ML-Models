```python
movies.dropna(inplace=True)
```

```python
#We want that all the genres are mentioned together like action,adventure,fantasy,scifi
#first we convert the string representing list of dictionaries to list of names extracted from those dictionaries # 36 mins

import ast
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
# We only want names of top three actors in the 'cast' column
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
     if i['job']=='Director':
        L.append(i['name'])
        break
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#Overview is a string so we will convert it into a list so that we can easily coordinate with the remaining columns (done in the next row)
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

WE NEED TO DO SOME TRANSFORMATION TO REMOVE THE SPACES BETWEEN THE TWO WORDS.FOR EXAMPLE: SAM WORTHINGTON SUCH THAT IT BECOMES SAMWORTHINGTON .
BUT WHY?
BECAUSE IF THE SPACE IS PRESENT THEN THE RECOMMENDER SYSTEM MAY CONSIDER IT AS '2 TAGS'. ALSO SAM MENDES IS PRESENT. SO THERE IS A POSSIBILITY THAT THE RECOMMENDER RECOMMENDS YOU SAM MENDES' MOVIES WHEN YOU WANTED TO WATCH SAM WORTHINGTON'S MOVIES.

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#create tags by cocatenating the different columns
movies['tags']= movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies["crew"]
```

```python
movies.head()
```

```python
# creating a new dataframe by removing unnecessary columns
new_df=movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
#convert the list in tags to string
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))
```

```python
#converting everything in tags to lowercase
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
!pip install NLTK  #(Reason Mentioned further)
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]

  for i in text.split():  # converting string to list
      y.append(ps.stem(i))

  return " ".join(y) #converting list to string
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
#Using countvectorizer for textVectorization process
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english') #we have taken max words 5000 (max_features)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() #.toarray for converting the sparse matrix to a numpy array
```

```python
cv.get_feature_names_out()
```

There are certain words like love,loved,loving etc which have the same meaning but are mentioned seperately. So to counter this problem , we will use steming and for that we will have to install NLTK library AND WE WILL MAKE CHANGES IN THE new_df Dataframe itself such that all the three words become love,love,love.

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0] #tells about the similarity of first movie with every other movie
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key= lambda x:x[1])[1:6]
# This piece of code first uses enumerate function because we need to sort the similarity score based on which movie is closest to
# the movie searched for. Now if we directly use sort toh score will get sorted and  we will lose the positions. for example avatar is the 1st
# movie and its similarity score will 1 with itself but if we directly apply sort then we will lose the position at which the similarity score is 1 . Thus, ENUMERATE is used
# next LIST function is used to make a list of enumerate scores. REVERSE true used because we want to print reverse list , in dec order
#  key  function is used because we need to sort on the basis of similarity score and not the index
# and [1:6] used because top 5 values need to be printed
```

```python
def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0] # calculates the index for a movie in new_df
  distances=similarity[movie_index] #calculating similarity score for a particular movie
  movies_list=sorted(list(enumerate(distances)),reverse=True,key= lambda x:x[1])[1:6]


  for i in movies_list:
    print(new_df.iloc[i[0]].title) #for fetching the title instead of indexes

```

```python
recommend('John Carter')
```
